```python
#Section A: Data Structures & Control Structures
#1 List Operations

#append
list=[1,2,3,4,5]
list.append(8)
print("append:",list)

list=[1,2,3,4,5]
list.extend([8,8])
print("extend:",list)

#insert
list=[1,2,3,4,5]
list.insert(0,1)
print("insert:",list)

#remove
list=[1,2,3,4,5]
list.remove(3)
print("remove:",list)

#pop
list=[1,2,3,4,5]
list.pop(4)
print("pop:",list)

#push
list=[1,2,3,4,5]
list.reverse()
print("reverse:",list)

#sorting
list=[7,21,9,2,29,24]
list.sort()
print("sort:",list)

#count
list=[8,12,3,4,5]
list.count(10)
print("count:",list)

#index
list=[2,4,6,9,10]
list.index(6)
print("index:",list)

#clear
number_list=list.copy()
number_list.clear()
print("clear:",number_list)
```

```python
# Section C: Exploring Pandas

import pandas as pd
import numpy as np

data = {
    "Name": ["Priya", "Rahul", "Sneha", "Kiran", "Anu","Pavan", "Teja", "Ravi", "Divya", "Sneha"],

    "Age": [21, 22, 23, 24, 25,26, 27, 28, 29, 23],

    "City": ["Hyderabad", "Vizag", "Chennai", "Delhi", "Mumbai","Pune", "Goa", "Bangalore", "Kolkata", "Chennai"],

    "Salary": [25000, 30000, np.nan, 40000, 35000,45000, np.nan, 50000, 55000, np.nan]
}

df = pd.DataFrame(data)
print("DataFrame:")
print(df)

# info()
print("DataFrame Info:")
print(df.info())

# describe()
print("Describe:")
print(df.describe())

# Select Name and City columns
print("Selected Columns:")
print(df[['Name', 'City']])

# Drop City column
df_drop = df.drop('City', axis=1)
print("After Dropping City Column:")
print(df_drop)

# Fill missing values 
mean_salary = df['Salary'].mean()
df['Salary'] = df['Salary'].fillna(mean_salary)
print("After Filling Missing Values:")
print(df)

# Remove duplicate rows
df = df.drop_duplicates()
print("After Removing Duplicate Rows:")
print(df)
```
