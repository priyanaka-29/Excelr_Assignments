```python
#Load the data sets

import numpy as np
import pandas as pd
patient=pd.read_csv("Patient_Data.csv")
patient
```

```python
billing=pd.read_csv("Billing_Data.csv")
billing
```

```python
#info() 
patient.info()
billing.info()
```

```python
# 2 Select relevant columns from billing
billing_columns=patient[['PatientID','Department','Doctor','BillAmount']]
billing_columns
```

```python
# 3.drop administrative columns

patient=patient.drop(['ReceptionistID','CheckInTime'],axis=1)
patient
```

```python
# 4. To find total bill amount per department using groupBy

department_billamount=patient.groupby('Department')['BillAmount'].sum()
department_billamount
```

```python
#5. Remove duplicate patient records based on PatientID

patient=patient.drop_duplicates(subset='PatientID')
patient                                                        

#---Drop Duplicate means remove the duplicates
```

```python
#6. Fill missing BillAmount values with the mean bill amount

billAmount_mean=patient['BillAmount'].mean()
billAmount_mean

#remove the null values and replaced them

patient.loc[:,'BillAmount'] = patient['BillAmount'].fillna(billAmount_mean)
patient

#We can replace the values using Fillna() funtcion 
#.loc[] is used to access particular rows and columns in pandas
```

```python
#7. Merge the billing dataset with patient dataset on PatientID

merge=pd.merge(patient, billing, on = 'PatientID', how='inner')
merge
```

```python
#8. Concatenate an additional DataFrame that contains new patients for the current week (row-wise)

new_patients=pd.DataFrame({'PatientID':[106,107],
                           'Department':['Cardiology','Oncologist'],
                           'Doctor':['Dr.Paul','Dr.Leo'],
                           'BillAmount':[6000,9000]
                          })
new_patients

#combined
Combined=pd.concat([patient ,new_patients], axis=0)
Combined
```

```python
#9. Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise)

new_columns = pd.DataFrame({ 'InsuranceCovered': ['Yes'] * len(patient), 'FinalAmount': patient['BillAmount'] * 0.9 }) 

final= pd.concat( [patient.reset_index(drop='True'), new_columns.reset_index(drop='True')], axis=1 ) 
final

#drop='True' means removes the old index values 
#axis=1 means columns side by side values

```

```python
Conclusion:
===========

1. dataset was cleaned and analysed successfully.
2. Missing bill amounts weere filled using mean values.
3. Duplicate patient records were removed.
4. Billing and patient datasets were merged successfully.
5. Groupby analysis helped identify department-wise revenue.
6. Concatenation was used to add new patient records and billing columns.
```
