```python
# load the dataset

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Cardiotocographic.csv")
df
```

```python
#1. Data Cleaning and Preparation:

#check missing values
df.isnull().sum()

===================
#The dataset was checked for missing values using isnull().sum()
No missing values were found in the dataset, so imputation or deletion techniques were not required.
```

```python
#check the data types

print(df.dtypes)


#The dataset in datatypes were checked using dtypes to that numerical columns were stored properly.
#No major inconsistencies in data types were found in the dataset.
#All numerical variables were already stored in appropriate numeric formats.
```

```python
#detech and treate outliers if necessary
df=df.drop_duplicates()
df

#statistical summary
print(df)
print(df.describe())
```

```python
# 2. Statistical Summary

#mean
mean=df.mean(numeric_only=True)

#median
median=df.median(numeric_only=True)

#standard deviation
standard_deviation=df.std(numeric_only=True)

#Interquatile range(IQR)
Q1=df.quantile(0.25, numeric_only=True)
Q3=df.quantile(0.75, numeric_only=True)
IQR=Q3-Q1

print("mean:", mean)
print("median:",median)
print("standard deviation:",standard_deviation)
print("Interquantile:",IQR)



#Summary:
==========
1. Some varibales show high variability based on standard deviation.
2. IQE values indicate the larger spread in the middle of 50% of data.
3. Certain variables may contain outliers or skewed distribution.
4. Variablitiy related fetal health features differ across observations.
```

```python
#3. Data visualizations

import matplotlib.pyplot as plt
import seaborn as sns

#1---------histogram
plt.figure(figsize=(5,6))
plt.hist(df['LB'])
plt.title("Histogram of numerical variables")
plt.show()

#-----------Boxplot
plt.figure(figsize=(5,6))
plt.boxplot(df['LB'])
plt.title("Boxplot of numerical variables")
plt.show()


'''
the datset mainly contains numerical variables and does not contain significant categorical values.
Therefore, bar charts and pie charts were not extensively used in this analysis

but i can use one column for categorical 'NSP' 
'''

#2--------------barcharts
nsp_count = df['NSP'].value_counts()
plt.figure(figsize=(6,5))
plt.bar(nsp_count.index, nsp_count.values)
plt.xlabel("NSP")
plt.ylabel("Count")
plt.title("Bar Chart of NSP")
plt.show()


#3------------Scatter plot
plt.figure(figsize=(6,7))
plt.scatter(df['LB'],df['AC'])
plt.xlabel("LB")
plt.ylabel("AC")
plt.title("Scatter plot")
plt.show()

#------------Correlation Heapmap
plt.figure(figsize=(12,8))
sns.heatmap( df.corr(numeric_only=True), annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

#df.corr(numeric_only=True) means calculate the correlation between numerical columns and it shows relationship between variables.
#annot=True means it displays the correlation values inside the each box.
#heapmap creates a colored correlation matrix
#cmap='ccolwarm' add colors to heapmap()


#4--------pair plot 
sns.pairplot(df)
plt.title("Pair plot")
plt.show()

#--------violin plot
plt.figure(figsize=(6,7))
sns.violinplot(data=df)
plt.xticks(rotation=90)
plt.title("Violin plot")
plt.show()

#pair plot ===> pair plot shows relationship between multiple numerical variables and it displays the scatter plot ans distributions and also it identify correlations,patterns, and clusters in dataset.
#violin plot===>It shows data distribution and it combines boxplot and histogram.
```
