```python
# excersize 1
#prime numbers

n=int(input("Enter Number:"))
count=0
for i in range(1,n+1):
    if n%i == 0:
        count=count+1
if count == 2:
    print("Number is Prime")
else:
    print("Number is not prime")
```

```python
#Excersize 2
#random number Product

import random
n1=random.randint(1,10)
n2=random.randint(1,10)
print("Enter n1 value:",n1)
print("Enter n2 value:",n2)

user_answer=int(input("Enter the product answer:"))
correct_answer=n1*n2

if user_answer == correct_answer:
    print("correct answer")
else:
    print("Wrong answer")
    print("Correct product answer is:",correct_answer)
```

```python
#excersize 4
#Hospital building system

room_charge_per_day = 2000
days = 4
doctor_fee = 1500
lab_test_cost = 300
number_of_tests = 3
medicine_charges = 2400

room_charges = room_charge_per_day * days
lab_charges = lab_test_cost * number_of_tests
total_bill = room_charges + doctor_fee + lab_charges + medicine_charges

if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0
final_bill = total_bill-discount

print("Room Charges:", room_charges)
print("Doctor Fee:", doctor_fee)
print("Lab Charges:", lab_charges)
print("Medicine Charges:", medicine_charges)
print("Total Bill Before Discount:", total_bill)
print("Discount:", discount)
print("Final Bill:", final_bill)
```
