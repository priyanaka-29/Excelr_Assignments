```python
# load the dataset 

import numpy as np
import pandas as pd
df=pd.read_csv("sales_data_with_discounts.csv")
df
```

```python
#Numerical columns identify
df.columns
df.head()
```

```python
df.isnull().sum()
```

```python
#calculate the mean, median, mode, and standard deviation 

#mean
int(df['Volume'].mean())
```

```python
#median
int(df['Volume'].median())
```

```python
#mode
int(df['Volume'].mode())
```

```python
#standard deviation
int(df['Volume'].std())
```

```python
Interpretations:

1.mean:
-->Mean represents the average value of the column.
-->It helps understand the overall central value of the data.
-->For example if the mean value is 5 then average value is 5.
      Mean=sum of values/no.of values 

2.Median:
-->Median is the middle value after sorting the data.
-->It is less affected by extreme value(like outliers).
-->If mean and median are very different, the data may be skewed.
    Even=middle of two numbers/2 after sorting
    odd=middle value after sorting

3.Mode:
-->Mode is the most frequently occuring value in the dataset.
-->It helps identify the commonly occuring values.
    mode=most repeated value

4.Standard Deviation
-->Standard deviation measures how much the values vary from the mean.
-->Low standard deviation then values are close to the average.
-->High standard deviation then values are spread out widely.
    
```

```python
#data visualizations 

import matplotlib.pyplot as plt
import seaborn as sns

#Histogram
plt.figure(figsize=(6,7))
plt.hist(df['Volume'])
plt.xlabel("Volume")
pl.ylabel("Frequency")
plt.title("Histogram")
plt.show()

#Box plot
plt.figure(figsize=(6,7))
plt.boxplot(df['Volume'])
plt.title("Box Plot")
plt.show()


#bar graphns
plt.figure(figsize=(6,7))
df['Brand'].value_counts().plot(kind='bar')
plt.title("Bar Graph")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.show()
```

```python
Inference / Analysis:
======================

Histogram Analysis:
--> Historgram shows distribution of Volume
--> If graph is not symmetric, data is skewed
-->Extreme values may indicate outliers

Boxplot Analysis:
-->Boxplot identifies the outliers
-->points outside whisker are extreme values
-->Box size represents interquartile range(IQR)

Bar Chart Analysis:
--> Bar Chart shows frequency of categories
-->Some brands may have higher sales frequency
-->Uneven bar heights indicate category imbalance
```

```python
Interpretation:
===============

1. Mean Represents the average value of the numerical columns
2. Median shows middle value of the dataset.
3. Mode indicates the most frequently occurring value.
4. Standard deviation measures how much the data varies from the mean.
5. Histogram shows the distribution and skewness of the data.
6. Boxplots helps identifies the outliers and extreme values.
7. Bar charts show the frequency distribution of categorical variable.


Conclusion:
==========

1. Descriptive analysis helps understand the overall behavior of the dataset.
2. Numerical statastics provide insights into average sales, variation, and frequent values.
3. Historgrams reveal data distribution patterns and skewness.
4. Boxplots identify outliers and unusual observations.
5. Bar charts helps analyze the frequency of categorical variable.
    
Finally, Data visualizations improves understanding before performing advanced analysis or machine learning.
```
